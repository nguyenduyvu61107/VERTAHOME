```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

try:
    df = pd.read_csv('/content/phong_tro_cleaned_v1 (1).csv')

    numeric_df = df.select_dtypes(include=['float64', 'int64'])

    corr_matrix = numeric_df.corr()

    plt.figure(figsize=(12, 10))
    sns.set_theme(style="white")

    heatmap = sns.heatmap(corr_matrix,
                          annot=True,
                          fmt=".2f",
                          cmap='coolwarm',
                          linewidths=0.5,
                          square=True,
                          cbar_kws={"shrink": .8})

    plt.title('Bản Đồ Tương Quan Dữ Liệu Giá Phòng Trọ', fontsize=16, pad=20)
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)

    plt.tight_layout()
    plt.show()

except FileNotFoundError:
    print("...")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('/content/phong_tro_cleaned_v1 (1).csv')

plt.figure(figsize=(10, 6))
sns.regplot(x='dien_tich', y='gia_thue', data=df, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title('Mối quan hệ giữa Diện tích và Giá thuê')
plt.show()

```

```python
# 3. Histogram: Phân phối giá thuê
plt.figure(figsize=(10, 6))
sns.histplot(df['gia_thue'], kde=True, color='purple')
plt.title('Biểu đồ phân phối giá thuê phòng trọ')
plt.show()
```
